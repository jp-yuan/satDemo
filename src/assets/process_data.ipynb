```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_rw_questions.csv")
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
# Keep the first occurrence of each id
df = df.drop_duplicates(subset=["question_id"], keep="first")

# Show the first few rows
df.head()
```

```python
import ast

df["choices"] = df["choices"].apply(ast.literal_eval)
```

```python
print(df["choices"].iloc[0])
# ['Choice A', 'Choice B', 'Choice C']
print(type(df["choices"].iloc[0]))
# <class 'list'>
```

```python
#shuffle

df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # shuffle
```

```python
import numpy as np

questions_per_test = 2 * 27
num_tests = int(np.ceil(len(df) / questions_per_test))

num_tests
```

```python
test_numbers = []
module_numbers = []

for i in range(len(df)):
    test_num = i // questions_per_test + 1  # test numbers start from 1
    module_num = (i % questions_per_test) // 27 + 1  # module 1 or 2
    test_numbers.append(test_num)
    module_numbers.append(module_num)

df['test_number'] = test_numbers
df['module_number'] = module_numbers

df
```
